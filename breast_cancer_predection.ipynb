```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop(df.columns[[-1,0]],axis= 1, inplace = True)
```

```python
df
```

```python
df['diagnosis'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
diag_map = {'M' :1, 'B' : 0}
```

```python
df['diagnosis'] = df['diagnosis'].map(diag_map)
```

```python
df
```

```python
x= df[['radius_mean','perimeter_mean', 'area_mean', 'concavity_mean', 'concave points_mean']]
```

```python
x
```

```python
y= df[['diagnosis']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
knn_y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(knn_y_pred, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression(random_state = 0)
```

```python
lr.fit(X_train, y_train)
```

```python
lr_y_pred = lr.predict(X_test)
```

```python
accuracy_score(lr_y_pred, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb =  GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
gnb_y_pred = gnb.predict(X_test)
```

```python
accuracy_score(gnb_y_pred, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy_all=[]
CVS_all=[]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
scores= cross_val_score(knn, x,y,cv = 10)
accuracy_all.append(accuracy_score(knn_y_pred, y_test))
CVS_all.append(np.mean(scores))

print("Accuracy:{0:2%}".format(accuracy_score(knn_y_pred, y_test)))
print("cross validation score:{0:2%} (+/- {1:2%})".format(np.mean(scores),np.mean(scores),np.std(scores)*2))
```
